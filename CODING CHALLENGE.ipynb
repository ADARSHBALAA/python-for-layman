```python
def print_rangoli(size):
    width = 4 * size - 3
    alphabets = [chr(96 + size - i) for i in range(size)]
    for i in range(size-1, -1, -1):
        row = '-'.join(alphabets[i:size])
        print(row.center(width, '-'))
    for i in range(1, size):
        row = '-'.join(alphabets[i:size])
        print(row.center(width, '-'))
N = int(input("Enter the size of the alphabet rangoli: "))
print_rangoli(N)
```

```python
n = int(input("Enter the size of the rangoli: "))
def rangoli(size):
    string='adarshbalan'
    list_1= []
    for i in range(size):
        line1 = '-'.join(string[i:size])
        list_1.append((line1[::-1] + line1[1:]).center(10 * size - 3, '-'))
    print('\n'.join(list_1[::-1] + list_1[1:]))

print(rangoli(n))
```

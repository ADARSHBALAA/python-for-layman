```python
class Account:
    def __init__(self, accnt_num, accnt_blnce, atm_pin):
        self.accnt = accnt_num
        self.blnce = accnt_blnce
        self.pin = atm_pin

    def withdraw(self, amount, atm_pin):
        if self.pin != atm_pin:
            print("Incorrect ATM pin. Withdrawal failed.")
        elif self.blnce - amount >= 0:
            print("You can withdraw the amount")
            self.blnce = self.blnce - amount
        else:
            print("Insufficient balance. Withdrawal failed.")

    def display(self):
        print(self.accnt, "your balance is", self.blnce)


if __name__ == "__main__":
    accnt_num1 = input("Enter your account number: ")
    pin1 = input("Enter your ATM pin: ")
    accnt1 = Account(accnt_num1, 10000, pin1)

    accnt_num2 = input("Enter your account number: ")
    pin2 = input("Enter your ATM pin: ")
    accnt2 = Account(accnt_num2, 10000, pin2)

    amount1 = float(input("Enter the amount to withdraw from account 1: "))
    atm_pin1 = input("Enter your ATM pin for account 1: ")
    accnt1.withdraw(amount1, atm_pin1)
    accnt1.display()

    amount2 = float(input("Enter the amount to withdraw from account 2: "))
    atm_pin2 = input("Enter your ATM pin for account 2: ")
    accnt2.withdraw(amount2, atm_pin2)
    accnt2.display()
```

```python
class Restaurant:
    def __init__(self):
        self.menu_items = {}         # Dictionary to store menu items
        self.booked_tables = set()   # Set to store booked table numbers
        self.customer_orders = {}    # Dictionary to store customer orders

    def add_item_to_menu(self, item_name, item_price):
        self.menu_items[item_name] = item_price

    def book_table(self, table_number):
        self.booked_tables.add(table_number)

    def customer_order(self, table_number, *items_ordered):
        if table_number not in self.customer_orders:
            self.customer_orders[table_number] = []
        for item in items_ordered:
            if item in self.menu_items:
                self.customer_orders[table_number].append(item)
            else:
                print(f"{item} is not available in the menu.")

    def print_menu(self):
        print("Menu:")
        for item, price in self.menu_items.items():
            print(f"{item}: ${price}")

    def print_booked_tables(self):
        print("Booked Tables:")
        for table in self.booked_tables:
            print(f"Table {table}")

    def print_customer_orders(self):
        print("Customer Orders:")
        for table, items in self.customer_orders.items():
            print(f"Table {table}: {', '.join(items)}")


# Usage example:
if __name__ == "__main__":
    # Create a restaurant instance
    my_restaurant = Restaurant()

    # Add items to the menu
    my_restaurant.add_item_to_menu("Pizza", 12.99)
    my_restaurant.add_item_to_menu("Burger", 8.99)
    my_restaurant.add_item_to_menu("Salad", 6.99)

    # Make table reservations
    my_restaurant.book_table(1)
    my_restaurant.book_table(3)

    # Take customer orders
    my_restaurant.customer_order(1, "Pizza", "Burger")
    my_restaurant.customer_order(3, "Salad", "Steak")
    my_restaurant.customer_order(5, "Pasta", "Coke")  # This table does not exist

    # Print the menu
    my_restaurant.print_menu()

    # Print table reservations
    my_restaurant.print_booked_tables()

    # Print customer orders
    my_restaurant.print_customer_orders()
```
